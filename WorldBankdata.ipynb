```python
# Select the necessary columns for analysis
df_filtered = data[["Country Name", "Series Name", "2020"]]

# Filter by series of interest (e.g. GDP per capita and employment)
df_filtered = df_filtered[df_filtered["Series Name"].isin(["GDP per capita (current US$)", "Employment to population ratio, 15+, female (%) (modeled ILO estimate)"])]

# Pivot the data to have one column per series of interest
df_pivoted = df_filtered.pivot(index="Country Name", columns="Series Name", values="2020")

#df_pivoted['Employment to population ratio, 15+, female (%) (modeled ILO estimate)'] = df_pivoted['Employment to population ratio, 15+, female (%) (modeled ILO estimate)'] / 100
df_pivoted
```

```python
import pandas as pd
import hvplot.pandas
import holoviews as hv

# Crear un gráfico de dispersión con los datos
scatter_plot = df_pivoted.hvplot.scatter(
    x="GDP per capita (current US$)",
    y="Employment to population ratio, 15+, female (%) (modeled ILO estimate)",
    title="GDP per capita vs. Female Employment",
    xlabel="Employment to population ratio, 15+, female (%) (modeled ILO estimate)",
    ylabel="GDP per capita (current US$)",
    hover_cols=["Country Name"]
)

# Cambiar el formato de las etiquetas de los ejes x
scatter_plot.opts(hv.opts.Scatter(xformatter='%.0f'))

scatter_plot
```

```python
# Select the necessary columns for analysis
df_filtered = data[["Country Name", "Series Name", "2020"]]

# Filter by series of interest (e.g. GDP per capita and employment)
df_filtered = df_filtered[df_filtered["Series Name"].isin(["GDP per capita (current US$)", "Employment to population ratio, 15+, male (%) (modeled ILO estimate)"])]

# Pivot the data to have one column per series of interest
df_pivoted = df_filtered.pivot(index="Country Name", columns="Series Name", values="2020")

df_pivoted['Employment to population ratio, 15+, female (%) (modeled ILO estimate)'] = df_pivoted['Employment to population ratio, 15+, male (%) (modeled ILO estimate)'] / 100
df_pivoted
```
