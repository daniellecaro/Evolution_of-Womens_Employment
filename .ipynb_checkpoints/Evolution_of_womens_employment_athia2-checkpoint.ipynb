# Data Preparation

```python
 # Initial imports
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
%matplotlib inline
```

```python
df.head()
```

```python
X = df.copy()
X.drop("annual_income", axis = 1, inplace = True)
X.head()
```

```python
#Define target vector
y = panel_df["annual_income"].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1) 
```

```python
#Creating standard scaler instance
scaler = StandardScaler()
```

```python
#fitting the StandardScaler 
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Fitting the Random Forest Model

```python
# Create the random forest classifier instance
rf_model = RandomForestRegressor(n_estimators = 300, random_state = 1)
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

# Making Predictions using the Random Forest Model

```python
predictions = rf_model.predict(X_test_scaled)
```

```python
# Random Forests in sklearn will automatically calculate the feature importance
importances = rf_model.feature_importances_

# We can sort the features by their importance
sorted(zip(rf_model.feature_importances_, X.columns), reverse = True)
```

```python
 # Visualize the features by importance
importances_df = pd.DataFrame(sorted(zip(rf_model.feature_importances_, X.columns), reverse=True))
importances_df.set_index(importances_df[1], inplace=True)
importances_df.drop(columns=1, inplace=True)
importances_df.rename(columns={0: 'Feature Importances'}, inplace=True)
importances_sorted = importances_df.sort_values(by='Feature Importances')
importances_sorted.plot(kind='barh', color='lightgreen', title= 'Features Importances', legend=False)
```

```python
# Reformat data of the independent variable X as a single column array
X = df1["annual_income"].values.reshape(-1, 1)

# Display sample data
X[:5]
```

```python
X = df1["annual_income"].to_frame()
X
```

```python
X = df1[["annual_income"]]
X
```

```python
# Create an array for the dependent variable y with the years experience data
y = df1["annual_income"]
```

```python
##Filter by gender, years of experience, and annual income 
df1_filtered = df1[["years_experience", "annual_income"]]
df1_filtered
```

# Building a Linear Regression Model

```python
# Create a model with scikit-learn
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# Display the slope
print(f"Model's slope: {model.coef_}")
```

```python
# Display the y intercept
print(f"Model's y intercept: {model.intercept_}")
```

```python
# Display the model's best fit line formula
print(f"Model's formula: y= {model.intercept_} + {model.coef_[0]}X")
```

```python
#Display the forumula to predict the annual income for a Male with 20 years of experience 
print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]} * 20")

#Predict the annual income for a Male with 20 years of experience 
y_20= model.intercept_ + model.coef_[0] * 20

#Display the prediction
print(f"Predicted annual income for a male employee with 20 years of experience: ${y_20:.2f}")
```

```python
predicted_y_values = model.predict(X)
```

```python
# Create a copy of the original data
df1_annual_income_predicted = df1_filtered.copy()

# Add a column with the predicted income values
df1_annual_income_predicted["annual_income_predicted"] = predicted_y_values

#Display the sample data
df1_annual_income_predicted.head()
```

```python
# Create a line plot of the predicted salary values 
best_fit_line = df1_annual_income_predicted.hvplot.line(
    x = "years_experience",
    y = "annual_income_predicted",
    color = "red")
best_fit_line
```

```python
panel_df_plot * best_fit_line
```

# Linear Regression Model Assessment 

```python
# Import relevant metrics 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Compute the metrics for the linear regression model 
score = model.score(X, y, sample_weight=None)
r2 = r2_score(y,predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)

# Print relevant metrics
print(f"The score is {score}")
print(f"The r2 score is {r2}")
print(f"The mse score is {mse}")
print(f"The rmse score is {rmse}")
print(f"The std is {std}")
```
