```python
for year in data.columns[4:]:
    data[year] = pd.to_numeric(data[year], errors='coerce')

# Verify the updated data types

print(data.dtypes)
```

```python
# Primero, derretimos ("melt") el DataFrame para que cada fila represente un único año-país-variable.
melted_data = data.melt(id_vars=["Country Name", "Country Code", "Series Name", "Series Code"], 
                        var_name="Year", 
                        value_name="Value")

# Luego, pivotamos el DataFrame derretido para que cada Serie sea una columna.
pivoted_data = melted_data.pivot_table(index=["Year", "Country Name", "Country Code"], 
                                       columns="Series Name", 
                                       values="Value").reset_index()

# Para hacer los nombres de las columnas más manejables, puedes quitar el nombre del nivel superior de las columnas.
pivoted_data.columns.name = None
pivoted_data
```

```python
pivoted_data["Year"] = pivoted_data["Year"].astype(int)
```

```python
pivoted_data.to_csv('Resources/newmap.csv', index = False)
```

```python
filtered_data = pivoted_data[pivoted_data["Year"] == 2019]
filtered_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filtered_data = pivoted_data[pivoted_data["Year"] == 2020].dropna(subset=["Employment in agriculture (% of total employment) (modeled ILO estimate)",
                                                                        "Employment in agriculture, female (% of female employment) (modeled ILO estimate)",
                                                                        "GDP per capita (current US$)"])

countries = filtered_data["Country Name"].unique()
num_countries = len(countries)

colormap = plt.cm.get_cmap("tab20", num_countries)  # Choose a colormap with enough distinct colors

for i, country in enumerate(countries):
    country_data = filtered_data[filtered_data["Country Name"] == country]
    plt.scatter(
        country_data["Employment in agriculture (% of total employment) (modeled ILO estimate)"],
        country_data["Employment in agriculture, female (% of female employment) (modeled ILO estimate)"],
        color=colormap(i),
        label=country,
        s=country_data["GDP per capita (current US$)"] * 50  # Adjust the multiplier for desired size range
    )
    # Agregar el código del país sobre cada punto
    for _, row in country_data.iterrows():
        plt.text(row["Employment in agriculture (% of total employment) (modeled ILO estimate)"],
                 row["Employment in agriculture, female (% of female employment) (modeled ILO estimate)"],
                 row["Country Code"],
                 fontsize=8,
                 ha='center',
                 va='bottom')

plt.xlabel("Employment in agriculture (% of total employment)")
plt.ylabel("Employment in agriculture, female (% of female employment)")
plt.title("Comparison of Female Employment in Agriculture (2020)")

plt.show()

```
